BREAST CANCER CLASSIFICATION USING PYTHON

In this model we will working on detection of breast cancer using Benign tumor and Malignant tumors .

Here Benign tumor is represented with value '1' and Malignant tumor is represented with value '0'

WORKFLOW :

Dataset Collection -> Dataset Preprocessing ->Training Dataset -> Logistic Regression Model -> Trained Logistic Regression Model -> New Data

Since we are dealing with Binary classification based data here we can use Logistic Regression which proves to be of a best fit

```python
# stage -1 : Importing packages / Dependencies
```

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# stage -2 : Data Collection and Preprocessing
```

```python
# loading data from sklearn
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset)
```

```python
# stage -3 : Loading data to a pandas Dataframe
```

```python
df=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
df.head()
```

```python
# stage -4 : Adding target Column to the dataframe
```

```python
df['label']=breast_cancer_dataset.target
df.head()
```

```python
# separating the Data Frame
```

```python
x=df.drop(columns='label',axis=1)
y=df['label']
print(x)
print(y)
```

```python
#stage -5 : Splitting the Dataset into train and test
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
print(x.shape,x_train.shape,x_test.shape)
```

```python
# stage -6 : Model Training
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
# stage -7 : Model Evaluation
```

```python
#training data
x_pred = model.predict(x_train)
accuracy=accuracy_score(y_train,x_pred)
print("Accuracy score =",accuracy)
```

```python
#test data
x_pred = model.predict(x_test)
accuracy=accuracy_score(y_test,x_pred)
print("Accuracy score =",accuracy)
```
